# Airbnb Feature Selection

`airbnb_preprocessing.ipynb`에서 저장한 전처리 결과를 불러와 피처 선택 실험을 진행한다.

```python
# 선형회귀용 피처 선택 시작용 복사본
linear_df = airbnb_df.copy()

# 공통 제외 컬럼
drop_cols = ['id', 'host_id', 'price']
drop_cols = [col for col in drop_cols if col in linear_df.columns]

linear_df = linear_df.drop(columns=drop_cols)

print(linear_df.columns.tolist())
print(linear_df.shape)
```

```python
categorical_cols = ['neighbourhood_group', 'room_type']
categorical_cols = [col for col in categorical_cols if col in linear_df.columns]

linear_df = pd.get_dummies(
    linear_df,
    columns=categorical_cols,
    drop_first=True,
    dtype=int
)

print(linear_df.columns.tolist())
print(linear_df.shape)
```

```python
selected_derived_features = [
    'night_crime_ratio_2km',
    'misdemeanor_ratio_2km',
    'felony_count_2km',
    'attraction_count_3km',
    'distance_to_city_center',
    'is_manhattan',
    'log_distance_to_nearest_station',
    'station_count_1km'
]

derived_feature_candidates = [
    'attraction_count_2km', 'attraction_count_3km',
    'distance_to_city_center', 'is_manhattan',
    'distance_to_nearest_station', 'log_distance_to_nearest_station',
    'station_count_300m', 'station_count_500m', 'station_count_1km',
    'log_station_count_300m', 'log_station_count_500m',
    'crime_count_0.5km', 'crime_count_1km', 'crime_count_2km',
    'felony_count_0.5km', 'felony_count_1km', 'felony_count_2km',
    'misdemeanor_count_0.5km', 'misdemeanor_count_1km', 'misdemeanor_count_2km',
    'night_crime_count_0.5km', 'night_crime_count_1km', 'night_crime_count_2km',
    'felony_ratio_0.5km', 'felony_ratio_1km', 'felony_ratio_2km',
    'misdemeanor_ratio_0.5km', 'misdemeanor_ratio_1km', 'misdemeanor_ratio_2km',
    'night_crime_ratio_0.5km', 'night_crime_ratio_1km', 'night_crime_ratio_2km'
]

drop_derived_features = [
    col for col in derived_feature_candidates
    if col in linear_df.columns and col not in selected_derived_features
]

linear_df = linear_df.drop(columns=drop_derived_features)

print("drop된 파생변수:", drop_derived_features)
print("남은 선택 파생변수:", [col for col in selected_derived_features if col in linear_df.columns])
```

```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
from sklearn.linear_model import LinearRegression
from statsmodels.stats.outliers_influence import variance_inflation_factor

# log_price 제외한 현재 후보 피처 전체
feature_cols = [col for col in linear_df.columns if col != 'log_price']

multicol_df = linear_df[feature_cols].copy()
multicol_df = multicol_df.replace([np.inf, -np.inf], np.nan).dropna()

# 1. 상관행렬
corr_matrix = multicol_df.corr(numeric_only=True).round(4)
display(corr_matrix)

# 2. VIF 계산
multicol_df_const = sm.add_constant(multicol_df)

vif_df = pd.DataFrame({
    'feature': multicol_df_const.columns,
    'VIF': [
        variance_inflation_factor(multicol_df_const.values, i)
        for i in range(multicol_df_const.shape[1])
    ]
})

vif_df = vif_df[vif_df['feature'] != 'const'].sort_values('VIF', ascending=False)
display(vif_df)

# 3. 각 변수를 나머지 변수들로 회귀한 R² / implied VIF
r2_rows = []

for target_feature in feature_cols:
    other_features = [col for col in feature_cols if col != target_feature]

    X = multicol_df[other_features]
    y = multicol_df[target_feature]

    model = LinearRegression()
    model.fit(X, y)

    r2 = model.score(X, y)
    implied_vif = 1 / (1 - r2) if r2 < 1 else np.inf

    r2_rows.append({
        'target_feature': target_feature,
        'R_squared': r2,
        'implied_vif': implied_vif
    })

r2_df = pd.DataFrame(r2_rows).sort_values('implied_vif', ascending=False)
display(r2_df)
```

```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
from sklearn.linear_model import LinearRegression
from statsmodels.stats.outliers_influence import variance_inflation_factor

feature_groups = {
    'location': [
        'attraction_count_3km',
        'distance_to_city_center',
        'neighbourhood_group_Manhattan'
    ],
    'crime': [
        'night_crime_ratio_2km',
        'felony_count_2km',
        'misdemeanor_ratio_2km'
    ],
    'station': [
        'station_count_1km',
        'log_distance_to_nearest_station'
    ],
    'room_type': [
        'room_type_Private room',
        'room_type_Shared room'
    ]
}

for group_name, features in feature_groups.items():
    valid_features = [col for col in features if col in linear_df.columns]
    group_df = linear_df[valid_features].copy()
    group_df = group_df.replace([np.inf, -np.inf], np.nan).dropna()

    print(f"\n===== {group_name.upper()} =====")

    # 1. 그룹 내부 상관계수
    print("\n[Correlation Matrix]")
    display(group_df.corr(numeric_only=True).round(4))

    # 2. 절편 포함 VIF
    print("\n[VIF]")
    group_df_const = sm.add_constant(group_df)

    vif_df = pd.DataFrame({
        'feature': group_df_const.columns,
        'VIF': [
            variance_inflation_factor(group_df_const.values, i)
            for i in range(group_df_const.shape[1])
        ]
    })

    vif_df = vif_df[vif_df['feature'] != 'const'].sort_values('VIF', ascending=False)
    display(vif_df)

    # 3. 각 변수를 나머지 변수들로 회귀했을 때 R² / implied VIF
    print("\n[R^2 and Implied VIF]")
    r2_rows = []

    for target_feature in valid_features:
        other_features = [col for col in valid_features if col != target_feature]

        if len(other_features) == 0:
            continue

        X = group_df[other_features]
        y = group_df[target_feature]

        model = LinearRegression()
        model.fit(X, y)

        r2 = model.score(X, y)
        implied_vif = 1 / (1 - r2) if r2 < 1 else np.inf

        r2_rows.append({
            'target_feature': target_feature,
            'R_squared': r2,
            'implied_vif': implied_vif
        })

    if r2_rows:
        r2_df = pd.DataFrame(r2_rows).sort_values('implied_vif', ascending=False)
        display(r2_df)
```
